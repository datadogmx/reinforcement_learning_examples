# Funciones de activación

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
def plot(call_back, interval = (1.5, 1.5), title =""):
    plt.ylim(interval)
    plt.locator_params(nbins=5)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.axhline(lw=1, c='black')
    plt.axvline(lw=1, c='black')
    plt.grid(alpha=0.4, ls='-.')
    plt.box(on=None)
    plt.title(title, fontsize=20)
    plt.plot(x, call_back(x), c='gray', lw=3)
    plt.savefig("activations/{}.png".format(title.lower().replace(" ","_").replace("ó","o")))
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
bipolar_sigmoid = lambda x: (1 - np.exp(-x)) / (1 + np.exp(-x))
arc_tan = lambda x: np.arctan(x)
relu = lambda x: list(map(lambda xx: max(xx, 0), x))
binary_step = lambda x: list(map(lambda xx: 1 if xx > 0 else 0, x))
```

## Gráficas

### Sigmoide

```python
plot(sigmoid, interval=(-0.4, 1.4), title="Sigmoide")
```

### Función Sigmoidal Bipolar

```python
plot(bipolar_sigmoid, interval=(-1.4, 1.4), title="Sigmoidal Bipolar")
```

### Función Arcotangente

```python
plot(bipolar_sigmoid, interval=(-1.4, 1.4), title="Arcotangente")
```

### Rectificador Lineal (ReLU)

```python
plot(relu, interval=(-0.4, 1.4), title="Rectificador Lineal")
```

###  Paso binario

```python
plot(binary_step, interval=(-0.4, 1.4), title="Paso Binario")
```
